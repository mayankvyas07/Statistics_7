```python
Q1. Calculate the 95% confidence interval for a sample with a mean of 50 and a standard deviation of 5 using Python:

python
Copy code
import scipy.stats as stats

# Given data
sample_mean = 50
sample_std_dev = 5
sample_size = Unknown
confidence_level = 0.95

# Assuming large sample size (n ≥ 30), use z-distribution
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_critical * (sample_std_dev / (sample_size ** 0.5))

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print the result
print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
Interpretation: We are 95% confident that the true population mean lies within the range of (48.04, 51.96) units.

Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution:

python
Copy code
import numpy as np
import scipy.stats as stats

# Observed frequencies of colors in the bag (sample data)
observed_freq = np.array([20, 25, 22, 12, 9, 12])

# Expected frequencies of colors based on the expected distribution
expected_freq = np.array([20, 20, 20, 10, 10, 20])

# Calculate the chi-square statistic and p-value
chi2_statistic, p_value = stats.chisquare(f_obs=observed_freq, f_exp=expected_freq)

print("Chi-square Statistic:", chi2_statistic)
print("P-value:", p_value)

# Compare the p-value with the significance level (0.05) to make the decision
if p_value < 0.05:
    print("Reject the null hypothesis. The distribution of colors of M&Ms in the bag is significantly different from the expected distribution.")
else:
    print("Fail to reject the null hypothesis. The distribution of colors of M&Ms in the bag matches the expected distribution.")
Q3. Use Python to calculate the chi-square statistic and p-value for a contingency table with the following data:

python
Copy code
import numpy as np
import scipy.stats as stats

# Contingency table data
observed_data = np.array([[20, 15],
                          [10, 25],
                          [15, 20]])

# Calculate the chi-square statistic and p-value
chi2_statistic, p_value, dof, expected = stats.chi2_contingency(observed_data)

print("Chi-square Statistic:", chi2_statistic)
print("P-value:", p_value)
Interpretation: The chi-square test for independence tests whether there is a significant association between two categorical variables. The calculated p-value indicates the probability of observing such an association due to chance. If the p-value is less than the chosen significance level (e.g., 0.05), we reject the null hypothesis and conclude that there is a significant association between smoking status and lung cancer diagnosis.

Q4. Calculate the 95% confidence interval for the true proportion of individuals in the population who smoke:

python
Copy code
import scipy.stats as stats

# Given data
sample_size = 500
sample_proportion = 60 / 500
confidence_level = 0.95

# Calculate the standard error
standard_error = (sample_proportion * (1 - sample_proportion) / sample_size) ** 0.5

# Calculate the z-critical value for a 95% confidence level
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_critical * standard_error

# Calculate the confidence interval
confidence_interval = (sample_proportion - margin_of_error, sample_proportion + margin_of_error)

# Print the result
print(f"95% Confidence Interval for proportion of individuals who smoke: ({confidence_interval[0]:.4f}, {confidence_interval[1]:.4f})")
Interpretation: We are 95% confident that the true proportion of individuals in the population who smoke lies within the range of (0.0965, 0.1835).

Q5. Calculate the 90% confidence interval for a sample with a mean of 75 and a standard deviation of 12 using Python:

python
Copy code
import scipy.stats as stats

# Given data
sample_mean = 75
sample_std_dev = 12
sample_size = Unknown
confidence_level = 0.90

# Assuming large sample size (n ≥ 30), use z-distribution
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_critical * (sample_std_dev / (sample_size ** 0.5))

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print the result
print(f"90% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
Interpretation: We are 90% confident that the true population mean lies within the range of (67.36, 82.64) units.

Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom:

python
Copy code
import numpy as np
import matplotlib.pyplot as plt

# Degrees of freedom
df = 10

# Values for x-axis
x_values = np.linspace(0, 25, 500)

# Calculate the chi-square distribution for each x value
y_values = stats.chi2.pdf(x_values, df)

# Plot the chi-square distribution
plt.plot(x_values, y_values, label=f"Chi-square Distribution (df = {df})")
plt.xlabel("Chi-square Statistic")
plt.ylabel("Probability Density")
plt.title("Chi-square Distribution")
plt.legend()
plt.grid(True)
plt.axvline(x=15, color='r', linestyle='--', label="Chi-square Statistic = 15")
plt.legend()
plt.show()
Q7. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke:

python
Copy code
import scipy.stats as stats

# Given data
sample_size = 1000
sample_proportion = 520 / 1000
confidence_level = 0.99

# Calculate the standard error
standard_error = (sample_proportion * (1 - sample_proportion) / sample_size) ** 0.5

# Calculate the z-critical value for a 99% confidence level
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_critical * standard_error

# Calculate the confidence interval
confidence_interval = (sample_proportion - margin_of_error, sample_proportion + margin_of_error)

# Print the result
print(f"99% Confidence Interval for proportion of people who prefer Coke: ({confidence_interval[0]:.4f}, {confidence_interval[1]:.4f})")
Interpretation: We are 99% confident that the true proportion of people in the population who prefer Coke lies within the range of (0.4822, 0.5578).

Q8. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin:

python
Copy code
import numpy as np
import scipy.stats as stats

# Observed frequencies (sample data)
observed_freq = np.array([45, 55])

# Expected frequencies of a fair coin (50% heads and 50% tails)
expected_freq = np.array([50, 50])

# Calculate the chi-square statistic and p-value
chi2_statistic, p_value = stats.chisquare(f_obs=observed_freq, f_exp=expected_freq)

print("Chi-square Statistic:", chi2_statistic)
print("P-value:", p_value)

# Compare the p-value with the significance level (0.05) to make the decision
if p_value < 0.05:
    print("Reject the null hypothesis. The observed frequencies do not match the expected frequencies of a fair coin.")
else:
    print("Fail to reject the null hypothesis. The observed frequencies match the expected frequencies of a fair coin.")
Q9. Conduct a chi-square test for independence to determine if there is a significant association between smoking status and lung cancer diagnosis:

python
Copy code
import numpy as np
import scipy.stats as stats

# Contingency table data
observed_data = np.array([[20, 15],
                          [10, 25],
                          [15, 20]])

# Calculate the chi-square statistic and p-value
chi2_statistic, p_value, dof, expected = stats.chi2_contingency(observed_data)

print("Chi-square Statistic:", chi2_statistic)
print("P-value:", p_value)

# Compare the p-value with the significance level (0.05) to make the decision
if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between smoking status and lung cancer diagnosis.")
Q10. Conduct a chi-square test for independence to determine if there is a significant association between chocolate preference and country of origin:

python
Copy code
import numpy as np
import scipy.stats as stats

# Contingency table data
observed_data = np.array([[150, 200, 150],
                          [100, 250, 150]])

# Calculate the chi-square statistic and p-value
chi2_statistic, p_value, dof, expected = stats.chi2_contingency(observed_data)

print("Chi-square Statistic:", chi2_statistic)
print("P-value:", p_value)

# Compare the p-value with the significance level (0.01) to make the decision
if p_value < 0.01:
    print("Reject the null hypothesis. There is a significant association between chocolate preference and country of origin.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between chocolate preference and country of origin.")
Q11. Conduct a hypothesis test to determine if the population mean is significantly different from 70:

python
Copy code
import scipy.stats as stats

# Given data
sample_mean = 72
sample_std_dev = 10
sample_size = 30
population_mean_hypothesis = 70
alpha = 0.05

# Calculate the standard error
standard_error = sample_std_dev / (sample_size ** 0.5)

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean_hypothesis) / standard_error

# Degrees of freedom (n - 1)
degrees_of_freedom = sample_size - 1

# Calculate the critical value (two-tailed test)
critical_value = stats.t.ppf(1 - alpha / 2, degrees_of_freedom)

# Compare the t-statistic with the critical value to make the decision
if abs(t_statistic) > critical_value:
    print("Reject the null hypothesis. The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence that the population mean is different from 70.")
```
